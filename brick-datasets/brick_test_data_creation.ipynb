```python
import time
import random
import pymysql

# 32 sensors with the data types you specified
SENSORS = {
    "memsvalue": "SMALLINT",
    "RsValue": "DECIMAL",
    "ppmvalue": "DECIMAL",
    "airQualityValue": "TINYINT",
    "airQualityLevel": "TINYINT",
    "Luxv": "TINYINT",
    "PM1.0Standard": "TINYINT",
    "PM2.5Standard": "TINYINT",
    "PM10Standard": "TINYINT",
    "PM1.0Atmospheric": "TINYINT",
    "PM2.5Atmospheric": "TINYINT",
    "PM10Atmospheric": "TINYINT",
    "voltage2": "DECIMAL",
    "mq2rsratio": "DECIMAL",
    "rsroratio2": "DECIMAL",
    "voltage3": "DECIMAL",
    "mq3rsratio": "DECIMAL",
    "rsroratio3": "DECIMAL",
    "voltage5": "DECIMAL",
    "mq5rsratio": "DECIMAL",
    "rsroratio5": "DECIMAL",
    "voltage9": "DECIMAL",
    "mq9rsratio": "DECIMAL",
    "rsroratio9": "DECIMAL",
    "o2percentage": "DECIMAL",
    "n2": "SMALLINT",
    "c2h5ch": "SMALLINT",
    "voc": "SMALLINT",
    "co": "SMALLINT",
    "co2": "SMALLINT",
    "tempreture": "DECIMAL",
    "humidity": "DECIMAL"
}

def generate_random_value(dtype):
    """
    Return a random value matching the approximate range/type 
    for the given MySQL column data type.
    """
    if dtype == "SMALLINT":
        # Signed SMALLINT range is -32768 to 32767; we'll pick a smaller range for demo
        return random.randint(0, 20000)
    elif dtype == "TINYINT":
        # Signed TINYINT range is -128 to 127; we'll pick 0-127 for demo
        return random.randint(0, 127)
    elif dtype == "DECIMAL":
        # We'll generate a random float up to 200 with 2 decimal places
        return round(random.uniform(0, 200), 2)
    else:
        return 0

def main():
    # Adjust connection parameters as needed
    connection = pymysql.connect(
        host='localhost',
        user='root',
        password='root',  # <-- replace with your password
        db='sensordb'     # <-- replace with your database name
    )

    device_id = 1  # Single sensor node

    try:
        with connection.cursor() as cursor:
            print("Starting to send sensor data for device_id = 1 in real time. Press Ctrl+C to stop.")
            
            while True:
                # Each cycle sends 5 partial batches (one per second)
                for _ in range(5):
                    partial_ts = int(time.time() * 1000)  # Unix epoch in milliseconds
                    # Insert data for the single device, each with 32 sensors
                    for sensor_name, dtype in SENSORS.items():
                        val = generate_random_value(dtype)
                        sql = """
                            INSERT INTO sensor_data_staging 
                                (device_id, sensor_name, partial_ts, reading_value)
                            VALUES (%s, %s, %s, %s)
                        """
                        cursor.execute(sql, (device_id, sensor_name, partial_ts, val))
                    connection.commit()
                    time.sleep(1)
                
                # At this point, 5 batches (with 5 different timestamps) have been sent
                # A trigger or stored procedure in MySQL should process these 5 batches
                # to compute an average timestamp and combine the 32 sensor readings into a single row.
                
    except KeyboardInterrupt:
        print("\nStopped sending data.")
    finally:
        connection.close()

if __name__ == "__main__":
    main()
```
